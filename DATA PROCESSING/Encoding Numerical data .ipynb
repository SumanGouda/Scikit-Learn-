```python
import requests
import pandas as pd
from datetime import datetime, timedelta

# Configuration
API_KEY = 'MZZVS2LHJYMT8B9QYW4SGRKEC'  # Replace with your key
LOCATION = "Bhanjanagar,Odisha,India"
START_DATE = (datetime.now() - timedelta(days=365)).strftime('%Y-%m-%d')  # 1 year back
END_DATE = datetime.now().strftime('%Y-%m-%d')

# Fetch historical data
url = f"https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/{LOCATION}/{START_DATE}/{END_DATE}?unitGroup=metric&include=days&key={API_KEY}&contentType=json"

response = requests.get(url)
data = response.json()

# Extract all available parameters
all_weather_data = []
for day in data['days']:
    record = {
        'date': day['datetime'],
        'temp_max': day['tempmax'],
        'temp_min': day['tempmin'],
        'temp': day['temp'],
        'feelslike': day['feelslike'],
        'humidity': day['humidity'],
        'dew': day['dew'],
        'precip': day['precip'],
        'precipprob': day['precipprob'],
        'snow': day['snow'],
        'windgust': day['windgust'],
        'windspeed': day['windspeed'],
        'winddir': day['winddir'],
        'pressure': day['pressure'],
        'cloudcover': day['cloudcover'],
        'visibility': day['visibility'],
        'uvindex': day['uvindex'],
        'sunrise': day['sunrise'],
        'sunset': day['sunset'],
        'conditions': day['conditions'],
        'description': day['description']
    }
    all_weather_data.append(record)

# Create DataFrame
df = pd.DataFrame(all_weather_data)

# Save to CSV
df.to_csv('bhanjanagar_weather_1year.csv', index=False)
print(f"Saved {len(df)} days of weather data to CSV")
print(df.head())
```
