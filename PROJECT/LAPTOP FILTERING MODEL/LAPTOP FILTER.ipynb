```python
# -------Load Modules --------
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt

# -------- Load Dataset---------
df = pd.read_csv('https://raw.githubusercontent.com/SumanGouda/Scikit-Learn-/refs/heads/main/PROJECT/LAPTOP%20FILTERING%20MODEL/laptops_enhanced.csv')
```

### Adding some extra colummns 

```python
# Extract processor type 
def extract_processor_suffix(text):
    if pd.isnull(text):         # ✅ Calls the pandas function to check if text is NaN or None
        return None
    
    match = re.search(r"(?:i\d|Ryzen\s\d)[\s\-]?\d{3,5}([A-Z])", text)
    if match:
        return match.group(1)
    return "Unknown"

df["Processor Series"] = df["Laptop"].apply(extract_processor_suffix)   
#df.rename(columns={"release_year" : "Release Year"}, inplace=True)
df.head()  
```
