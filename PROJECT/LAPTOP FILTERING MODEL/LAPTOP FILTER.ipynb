### Adding some extra colummns 

```python
# Extract processor type 
def extract_processor_suffix(text):
    if pd.isnull(text):         # âœ… Calls the pandas function to check if text is NaN or None
        return None
    
    match = re.search(r"(?:i\d|Ryzen\s\d)[\s\-]?\d{3,5}([A-Z])", text)
    if match:
        return match.group(1)
    return "Unknown"

df["Processor Suffix"] = df["Laptop"].apply(extract_processor_suffix)   
df.rename(columns={"release_year" : "Release Year"}, inplace=True)
df.head()  
```
